```python
import numpy as np

class BGDRegressor:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0
        
        for _ in range(self.epochs):
            y_hat = np.dot(X, self.coef_) + self.intercept_
            
            error = y - y_hat
            
            d_intercept = -2 * np.mean(error)
            d_coef = -2 * np.dot(X.T, error) / n_samples
            
            self.intercept_ -= self.learning_rate * d_intercept
            self.coef_ -= self.learning_rate * d_coef

    def predict(self, X):
        return np.dot(X, self.coef_) + self.intercept_
```

```python
BGD = BGDRegressor(learning_rate=0.1, epochs=1000)
BGD.fit(X_train_scaled, y_train)

y_pred = BGD.predict(X_test_scaled)

from sklearn.metrics import r2_score
print(f'The R2_score of BATCH_GRADIENT_DECENT is : {r2_score(y_test, y_pred)}')
```

**STOCHASTIC_GRADIENT_DECENT**

```python
class SGDRegressor:
    def __init__(self,learning_rate=0.1,epochs=100):
        self.coef_ = 0
        self.intercept = 0
        self.lr = learning_rate
        self.ep = epochs

    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.ep):
           for j in range(X_train.shape[0]):
               idx = np.random.randint(0,X_train.shape[0])
               y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
               int_der = -2*(y_train[idx] - y_hat)
               self.intercept_ = self.intercept_ - (self.lr * int_der)
               
               coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
               self.coef_ = self.coef_ - (self.lr * coef_der)

                

        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_ 
```

```python
SGD = SGDRegressor(learning_rate=0.0001, epochs=10)
SGD.fit(X_train_scaled, y_train)

y_pred = SGD.predict(X_test_scaled)
print(f'The R2_score of STOCHASTIC_GRADIENT_DECENT is : {r2_score(y_test, y_pred)}')
```

```python
class MBGDRegressor:
    def __init__(self,batch_size=32,learning_rate=0.001,epochs=100):
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.epochs = epochs

    def fit(self,X_train,y_train):
        self.intercept_= 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):

            indices = np.random.permutation(X_train.shape[0])
            X_train = X_train[indices]
            y_train = y_train[indices]

            for j in range(0, X_train.shape[0], self.batch_size):

                X_batch = X_train[j:j+self.batch_size]
                y_batch = y_train[j:j+self.batch_size]

                y_hat = np.dot(X_batch,self.coef_) + self.intercept_
                error = y_batch - y_hat

                in_der = -2 * np.mean(error)
                co_der = -2 * np.dot(X_batch.T,error) / len(X_batch)

                self.intercept_ -= self.learning_rate * in_der
                self.coef_ -= self.learning_rate * co_der

    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```
