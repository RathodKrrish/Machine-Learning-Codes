```python
from sklearn.datasets import load_diabetes

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y = load_diabetes(return_X_y = True)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2,random_state = 12)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
print(f'the model coefficient is : {model.coef_}')
print(f'the model intercept is : {model.intercept_}')
```

**NOW LETS MAKE A CLASS TO FIND THE DERIVATIVES OF THESE**

```python
class BGDRegressor:
    def __init__(self,learning_rate=0.1,epochs=100):
        self.coef_ = 0
        self.intercept = 0
        self.lr = learning_rate
        self.ep = epochs

    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.ep):
            # update all the coef and the intercept
            y_hat = np.dot(X_train,self.coef_) + self.intercept_
            #print("Shape of y_hat",y_hat.shape)
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)
            
            coef_der = -2 * np.dot((y_train - y_hat),X_train)/X_train.shape[0]
            self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_ 
```

```python
bgdr = BGDRegressor(0.5,1000)
```

```python
bgdr.fit(X_train,y_train)
```

```python
y_pred = bgdr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
